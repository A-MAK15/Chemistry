```python
#Task 2 : Data Preprocessing
df.info() #Checking for null values => All our rows are filled
```

```python
X = df.loc[:, df.columns != "Wine"] #Extracting our features from the dataset
y =  df.loc[:, df.columns == "Wine"] #Exctracting our target from the dataset
# print(X)
# print(y)
```

```python
x_scaler = MinMaxScaler() #Variable to scale our features
normalized_inputs = x_scaler.fit_transform(X) #As can be seen in the data visualization of averages for these attributes, Proline conists of
#a higher number in comparison to other attributes and therefore all our features will need to be standardised so that Proline does not dominate and affect the accuracy of our model.
X = normalized_inputs
```

```python
#MLP
mlp_x_main, mlp_x_val, mlp_y_main, mlp_y_val = train_test_split(X, y, test_size=0.25, random_state=42)#Providing 25% of the data for validation
mlp_x_train, mlp_x_test, mlp_y_train, mlp_y_test = train_test_split(mlp_x_main, mlp_y_main, test_size=0.25, random_state=42)#Providing 25% of the data for testing and the remaining 50% for training the data
#Random Forest
rf_X_train, rf_X_test, rf_y_train, rf_y_test = train_test_split(X, y,test_size=0.50, random_state=42)
```

```python
#Task 3 : Model Selection
#Multilayer perceptron
MLP = MLPClassifier(hidden_layer_sizes=(20), max_iter=2000, alpha=0.0001, verbose=True, random_state=42)# I chose the MLP classifier algorithm for the model selection
# MLP.fit(x_main, y_main)
MLP.fit(mlp_x_train, mlp_y_train.values.ravel())
mlp_y_pred = MLP.predict(mlp_x_test)
```

```python
#Random Forest
rfc = RandomForestClassifier(criterion="entropy", max_depth=8, min_samples_split=10, random_state=42, verbose=True)
rfc.fit(rf_X_train, rf_y_train.values.ravel())
output_pred = rfc.predict(rf_X_test)
```

```python
#Task 4 : MLP Evaluation
print("---------------------------------MLP Algorithm Start---------------------------------------------------")
print(confusion_matrix(mlp_y_test, mlp_y_pred))
print()
print("Accuracy Score")
print(accuracy_score(mlp_y_test, mlp_y_pred))
print()
print("Classification report")
print(classification_report(mlp_y_test, mlp_y_pred))
print("Confusion matrix")
sns.heatmap(confusion_matrix(mlp_y_test, mlp_y_pred), annot=True)
print()
```

```python
#Random Forest Model Evaluation
print("---------------------------------Random Forest Algorithm Start---------------------------------------------------")
print(confusion_matrix(rf_y_test, output_pred))
print()
print("Accuracy Score")
print(accuracy_score(rf_y_test, output_pred))
print()
print("Classification report")
print(classification_report(rf_y_test, output_pred))
print("Confusion matrix")
sns.heatmap(confusion_matrix(rf_y_test, output_pred), annot=True)
```

```python
#ROC Curve
def plot_roc_curve(model, X_test, y_test, label):
    probs = model.predict_proba(X_test)
    fpr, tpr, _ = roc_curve(y_test, probs[:, 1], pos_label=1)
    plt.plot(fpr, tpr, label=label)

plt.figure(figsize=(10, 15))
plot_roc_curve(MLP, mlp_x_test, mlp_y_test, "MLP ROC Curve")
plot_roc_curve(rfc, rf_X_test, rf_y_test, "RF ROC Courve")
plt.title('ROC Curves for MLP and RF')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.show()
```

```python
#Task 5
print("MLP Accuracy :",accuracy_score(mlp_y_test, mlp_y_pred))
print("Random Forest Accuracy :",accuracy_score(rf_y_test, output_pred))
print("Difference :",accuracy_score(rf_y_test, output_pred) - accuracy_score(mlp_y_test, mlp_y_pred))
#The Random Forest Algorithm is performing better than the MLP Algorithm because the Random Forest consists of a hidden 
# layer that has about 20 nodes.
```
